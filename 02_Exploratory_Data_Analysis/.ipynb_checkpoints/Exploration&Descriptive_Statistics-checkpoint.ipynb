## Data transformation: Skills

```python
# Regex test
if re.search(r",", df.skills[0]):
    print('success')
```

```python
# transform str list into list

def transform_list(df):
    for index, row in df.skills.dropna().iteritems():

        if re.search(r",", df.skills[index]):
            try:
                df.skills[index] = ast.literal_eval(df.skills[index]) 
            except:
                #print('Error', index, row)
                continue

        else:
            #print(index,row)
            continue
    return df
```

```python
# function to multiprocess
from multiprocessing import  Pool

def parallelize_dataframe(df, func, n_cores=4):
    df_split = np.array_split(df, n_cores)
    pool = Pool(n_cores)
    df = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df
```

```python
df_clean = parallelize_dataframe(df,transform_list)
```

```python
# save as json
df_clean.to_json("data_clean.json")
```

```python
# Collect all unique skills in a list

list_of_skills = []

for index, row in df.skills.iteritems():
    
    if  isinstance(row,list):
        
        for skill in row:
            if skill not in list_of_skills:
                list_of_skills.append(skill)
                #print("success", skill, len(list_of_skills))
            else:
                pass
    
    if  isinstance(row,str):
        
        if row not in list_of_skills:
                list_of_skills.append(row)
                #print("success", row, len(list_of_skills))
        else:
                pass
    else:
        #print(row, type(row))
        pass   
```

```python
# Save list of skills
import pickle

filename = 'list_of_skills'
outfile = open(filename,'wb')
pickle.dump(list_of_skills,outfile)
outfile.close()
```

```python
# Load list
import pickle
filename = 'list_of_skills'
infile = open(filename,'rb')
list_of_skills = pickle.load(infile)
infile.close()
```

```python
# Count how many times each skill occurs

skill_dict = { i : 0 for i in list_of_skills }

for index, row in df.skills.iteritems():
    
    if  isinstance(row,list):
        
        for skill in row:
            skill_dict[skill] += 1
    
    if  isinstance(row,str):
        
        skill_dict[row] += 1

    else:
        pass   
```

```python
# Save dict of skill count
import pickle

filename = 'dict_of_skills'
outfile = open(filename,'wb')
pickle.dump(skill_dict,outfile)
outfile.close()
```

```python
# Load dict
import pickle
filename = 'dict_of_skills'
infile = open(filename,'rb')
skill_dict = pickle.load(infile)
infile.close()
```

```python
# Top 10 skills in terms of frequency
skill_count = pd.DataFrame.from_dict(skill_dict, orient='index',columns=['count'])
skill_count = skill_count.sort_values("count", ascending=False)
skill_count.sort_values("count", ascending=False).head(10)
```

```python
#Exclude all skills that only occure once
skill_columns = skill_count[skill_count['count'] > 1].index

skill_columns = list(skill_columns[0:10])

# add skills as columns to dataframe and populate it

for i in skill_columns:
    df[i]= 0 

for skill in skill_columns:
    
    for index, row in df.skills.dropna().iteritems():
        
        if skill in row:
            
            #df.loc[index,(skill)]=1
            df[skill][index]=1

    print(skill, 'is done.')
```

# 1.1 Wages over time 

## Preparing the time dimension

```python
# Create date_time object
df["date_3"] = pd.to_datetime(df.date_2)
# As datetime object
df['month_year'] = pd.to_datetime(df['date_3']).dt.to_period('M')
```

```python
# overall median wage by month
monthly = df.groupby(['month_year']) \
       .agg({'date':'size',"rate":"median"}) \
       .rename(columns={'date':'count', 'rate':'median_wage'})
```

```python
# Count per month all categories (date as string)

# Create month_year column
df['month_year'] = pd.to_datetime(df['date_3']).dt.to_period('M')
df['month_year'] = df['month_year'].values.astype(str)


monthly_median = df.groupby(['month_year']) \
       .agg({'date':'size',"rate":"median"}) \
       .rename(columns={'date':'count', 'rate':'median_wage'})

ax = sns.lineplot(x = monthly_median.index, y=monthly_median.median_wage)
ax.set_xticklabels(rotation=90, labels=monthly_median.index)
plt.savefig('wage_over_time.png',dpi=300,bbox_inches = "tight")
```

```python
# Wage by category over time

monthly = df.groupby(['category_1','month_year'],as_index=False) \
       .agg({'date':'size',"rate":"median"}) \
       .rename(columns={'date':'count', 'rate':'median_wage'})
monthly.head(10)
```

```python
ax = sns.lineplot(x='month_year', y='median_wage', hue='category_1', 
             data=monthly[monthly["count"] > 10].sort_values("median_wage", ascending=False))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) # put legend outside
ax.set_xticklabels(rotation=90, labels=monthly.month_year)
plt.xlabel("Month")
plt.ylabel("Median wage in $")
plt.savefig('wage_by_category.png',dpi=300,bbox_inches = "tight")
```
